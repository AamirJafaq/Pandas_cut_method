```python
import pandas as pd
```

```python
data=pd.DataFrame({"person": ["p1", "p2","p3", "p4", "p5", "p6", "p7", "p8", "p9", "p10"], "age": [30, 32, 45, 24, 55, 65, 50, 70, 66, 17]})
```

```python
data
```

# What does pd.cut method do in pandas 

By using pd.cut() method, we can create categorical columns by cutting down the values into different classes or bins.

```python
categorical_col=pd.cut(data.age, bins=3, labels=["young","medium", "old"]) #here we divide the age into 3 bins of equal length.
```

```python
categorical_col
```

```python
data["age_cat"]=categorical_col
```

```python
data
```

```python
#here we separate the data into desired bins.
categorical_col0=pd.cut(data.age, bins=[data.age.min(), 35, 60, data.age.max()], labels=["short","medium", "long"])
```

```python
data["age_cat0"]=categorical_col0
```

```python
data
```

The "pd.cut()" method does not unclude the lowest value of bins. So, it shows a null value in the dataframe. For that, we need to include include_lowest parameter.

```python
categorical_col1=pd.cut(data.age, bins=[data.age.min(), 35, 60, data.age.max()], labels=["short","medium", "long"], include_lowest=True)
```

```python
data["age_cat1"]=categorical_col1
```

```python
data
```
